#### Manual Simple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 5, 4, 5]
}
df = pd.DataFrame(data)
df
```

```python
x = df["x"]
y = df["y"]

x_mean = x.mean()
y_mean = y.mean()
print(x_mean)
print(y_mean)
```

```python
numerator = sum((x - x_mean)*(y - y_mean))
denominator = sum((x - x_mean)**2)
beta_1 = numerator / denominator
beta_1
```

```python
beta_0 = y_mean - beta_1 * x_mean
beta_0
```

```python
df["y_hat"] = beta_0 + beta_1 * df["x"]
df
```

```python
plt.scatter(df["x"], df["y"], color="blue", label="Data points")
plt.plot(df["x"], df["y_hat"], color="red", label="Regression line")
plt.legend()  # This shows the legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.grid(True)
plt.show()  # This prevents the matplotlib object output (or semi-colon after plt.plot();)
```

#### Simple Linear Regression with Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["x"]] # Must be 2D (n samples x 1 feature)
y = df["y"] # Must be 1D (n samples)
```

```python
model = LinearRegression() # This creates a blank Linear Regression model. It hasn’t learned anything yet.
model.fit(X, y) # This is the learning step.
# It does the same thing we did manually:
# Minimizes RSS
# Finds the best β₀ and β₁
# Stores them inside the model
```

```python
sk_beta_0 = model.coef_[0]
sk_beta_1 = model.intercept_
print("sk_beta_0:", sk_beta_0)
print("sk_beta_1:", sk_beta_1)
```

```python
df["y_hat_sk"] = model.predict(X)
df
```

```python
plt.scatter(df["x"], df["y"], color="blue", label="Data Points")
plt.plot(df["x"], df["y_hat_sk"], color="red", label="Sklearn Regression Line")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression with Sklearn")
plt.grid(True)
plt.show()
```
